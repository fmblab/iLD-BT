# Generate domain

**LD-BT generate-domain step.** Build the combinatorial design space for a target motif: every sequence obtained by substituting the chosen alphabet at the variable positions of the wild-type motif, all other positions held fixed. Domain size is derived from the inputs (`len(alphabet) ** n_positions`), not hardcoded.

Edit the **Parameters** cell, then Run All. Output: `data/<name>/domain.csv`.

**Figures:** the enumerated design space underlies library design (Fig 2) and the sequence landscapes (Fig 3, Fig 5, Fig 7).

```python
import os, sys
ROOT = os.path.abspath(os.getcwd())
SRC = os.path.join(ROOT, 'src')
if SRC not in sys.path:
    sys.path.insert(0, SRC)
print('repo root :', ROOT)
print('src on path:', SRC in sys.path)
```

```python
from utils import WT_SEQ, STD_AAS, generate_all_combos
print('imported generate_all_combos; default WT =', WT_SEQ)
```

## Parameters
The published SrUGT76G1 A1-helix design space.

```python
NAME      = 'SrUGT76G1'
WT        = WT_SEQ                 # wild-type motif
POSITIONS = [3, 4, 5, 6, 8]        # 0-based variable positions
ALPHABET  = STD_AAS                # 20 standard amino acids
OUTPUT    = os.path.join('data', NAME, 'domain.csv')
print('name      :', NAME)
print('positions :', POSITIONS)
print('alphabet  :', ALPHABET, '(%d letters)' % len(ALPHABET))
```

## Validate positions

```python
for pos in POSITIONS:
    assert 0 <= pos < len(WT), 'position %d out of range for motif length %d' % (pos, len(WT))
print('all', len(POSITIONS), 'positions in range for motif length', len(WT))
```

## Build the domain

```python
combos = generate_all_combos(WT, POSITIONS, ALPHABET)
print('domain size : {:,}  (= {} ** {})'.format(len(combos), len(ALPHABET), len(POSITIONS)))
print('first 3     :', combos[:3])
```

## Write `domain.csv`

```python
os.makedirs(os.path.dirname(OUTPUT), exist_ok=True)
with open(OUTPUT, 'w', encoding='utf-8') as fh:
    fh.write('index,Candidate_Sequence\n')
    for i, seq in enumerate(combos):
        fh.write('%d,%s\n' % (i, seq))
print('written to :', OUTPUT, '(%d rows)' % len(combos))
```
